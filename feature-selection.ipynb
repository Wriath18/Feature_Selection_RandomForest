```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
data = pd.read_csv("/kaggle/input/mobile-price-classification/train.csv")
```

```python
data.head()
```

```python
X = data.iloc[:,0:-1]
Y = data.iloc[:,-1]
print(X.head())
print(Y.head())
```

```python
Selector = SelectKBest(score_func = chi2, k=10)
fitting = Selector.fit(X,Y)
```

```python
print(fitting.scores_)
dfscores = pd.DataFrame(fitting.scores_)
Feature_names = pd.DataFrame(X.columns)
```

```python
Feature_names = pd.concat([Feature_names,dfscores], axis=1)
```

```python
Feature_names.columns = ["Features", "Scores"]
```

```python
# Feature_names.sort_values(by="Scores", ascending=False)
TenFeatures = Feature_names.sort_values(by="Scores", ascending=False).head(10)["Features"].values
```

Feature Importance

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
model = RandomForestClassifier()
model.fit(X,Y)
```

```python
model.feature_importances_
```

```python
importances = pd.DataFrame(model.feature_importances_, index=X.columns, columns=[
    "Importance"
])
```

```python
importances
```

```python
importances = importances.sort_values(by="Importance", ascending=False)
```

```python
plt.figure(figsize=(20,10))
plt.bar(importances.index, importances["Importance"])
```
